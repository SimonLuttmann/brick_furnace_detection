```python
mask_dir = '/mnt/c/Users/arneh/GitHub/brick_furnace_detection/data/Brick_Data_Train/Mask'
mask_files = [f for f in os.listdir(mask_dir) if f.lower().endswith('.tif')]
mask_images = [tiff.imread(os.path.join(mask_dir, f)) for f in mask_files[:100]]
```

```python
import matplotlib.pyplot as plt
tif_image1 = tif_images[1]
rgb_img = np.clip(tif_image1[:, :, [2, 1, 0]] * 3, 0, 1)
mask_image1 = mask_images[1]

plt.figure(figsize=(16, 8))
plt.subplot(1, 2, 1)
plt.imshow(rgb_img)
# Overlay mask: where mask_image1 > 0, show in red with some transparency
plt.imshow(np.ma.masked_where(mask_image1 <= 0, mask_image1), cmap='Reds')
plt.title('RGB Image with Mask Overlay')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(rgb_img)
plt.title('RGB Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
x = 10  # Number of images to plot

for i in range(x):
    img = tif_images[i]
    mask = mask_images[i]
    rgb = img[:, :, [2, 1, 0]]  # BGR to RGB

    plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1)
    plt.imshow(np.clip(rgb, 0, 1))
    plt.imshow(np.ma.masked_where(mask <= 0, mask), cmap='Reds', alpha=0.4)
    plt.title(f'Image {i} with Mask Overlay')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(np.clip(rgb, 0, 1))
    plt.title(f'Image {i} RGB')
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```
